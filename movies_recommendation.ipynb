```python
import numpy as np
import pandas as pd
```

```python


credits = pd.read_csv('tmdb_5000_credits.csv', error_bad_lines=False)

movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.head()
```

```python
movies.shape
```

```python
credits.head()
```

```python
movies = movies.merge(credits, on= 'title')       # merging movies and credits datasset on title column
movies.head(2)
```

```python
movies.info()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]    #dropping unnecessary columns from dataset
movies.head()
```

```python
movies.isna().sum()
```

```python
movies = movies.dropna()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'].values
```

```python
def convert2(obj):
    L =[]
    conv = 0
    for i in ast.literal_eval(obj):
        if conv != 3:
            L.append(i['name'])
            conv+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply( lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply( lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply( lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply( lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords']+ movies['cast'] + movies['crew']   #concatinating overview,genres,keywords,cast and crew column to crate tags column
```

```python
movies = movies[['movie_id','title','tags']]     #dropping extra column from movies dataset
movies.head()
```

```python
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
```

```python
movies['tags'][0]
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
cv.get_feature_names_out()[:100]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    A = []
    for i in text.split():
        A.append(ps.stem(i))
    return " ".join(A)
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
cv.get_feature_names_out()[:100]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]       #gives index of movie
    distances = similarity[movie_index]                           #gives distance between vectors
    movies_list = sorted(list(enumerate(distances)),reverse=True,key= lambda x:x[1])[1:6]    #will give top 5 movies which has identical similarity values
    print('ID','Movies' )
    for i in movies_list:
        print(i[0],movies.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
